```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

```

```python
def show_info(soil_table, k_species :int):
    print("Species number")
    print(k_species)
    print(soil_table.shape)
    print("Mean Vector")
    print(np.round(soil_table.mean(axis=0),2))
    print("Standard deviation vector")
    print(np.round(soil_table.std(axis=0),2))
    sns.distplot(soil_table[:,k_species])
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5) 
```
