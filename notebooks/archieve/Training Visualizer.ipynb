```python
import pandas as pd
from plot_utils import  *
import numpy as np

%matplotlib widget
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
def plot_loss(epochs, record, n_samples:int, title:str):
    string_append = " ; Epochs: "+str(epochs[-1])+" ;Traning Samples "+str(n_samples)
    plt.plot(epochs, record)
    plt.title(title + string_append)
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    fig = plt.gcf()
    fig.set_size_inches(18.5, 10.5)
```

```python
results_path = "../data/plot4/21502_3_data"

discriminator_loss = pd.read_csv(results_path+"/_disc_loss.csv",header=None ).values.flatten()
generator_loss = pd.read_csv(results_path+"/_gen_loss.csv",header=None ).values.flatten()

epoch_record = pd.read_csv(results_path+"/_epoch_record.csv",header=None ).values.flatten()

temp =  pd.read_csv(results_path+"/_training.csv",header=None ).values.flatten()

total_epochs, total_iters = temp
del temp

training_indices =  pd.read_csv(results_path+"/training_indices.csv",header=None ).values.flatten()

N_samples = training_indices.shape[0]
```

```python
plot_loss(epoch_record, generator_loss, N_samples, "Generator Loss")
```

```python
plot_loss(epoch_record, discriminator_loss, N_samples, "Generator Loss")
```
