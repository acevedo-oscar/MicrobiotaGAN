```python
import pandas as pd
from plot_utils import  *
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#k1_gan = pd.read_csv('gans_jsd_record/original.csv')
#k2_gan = pd.read_csv('gans_jsd_record/k2_2000_old.csv')

k2_gan = pd.read_csv('gans_jsd_record/low_lr_k2_2000_epochs.csv')
k3_gan = pd.read_csv('gans_jsd_record/low_lr_k3_2000_epochs.csv')
k4_gan = pd.read_csv('gans_jsd_record/low_lr_k4_2000_epochs.csv')
```

# K3 y K4 (nuevas)

```python

plot_gan_curve(k2_gan, "K2 2000 epochs", debug_flag=False)
plot_gan_curve(k3_gan, "K3 2000 epochs", debug_flag=False)
plot_gan_curve(k4_gan, "K4 2000 epochs", debug_flag=False)

plt.ylabel("JSD(GAN,Test)",fontsize=40)
plt.xlabel("#Samples used to train",fontsize=40)
plt.title("GAN JSD - Samples",fontsize=40)

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.savefig("jsd_k_dirichlets.png", dpi=300)
```

```python
sub =   k4_gan['id_sample_N'] == 15255
sub2 =   k4_gan['id_sample_N'] == 23640
final = np.logical_or(sub2  , sub)
fast_k4 = k4_gan[final]
```
